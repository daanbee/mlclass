```python
# Task 1
import pandas as pd

odd_numbers = [i for i in range(1, 101) if i % 2 != 0]
print(odd_numbers)


array_dict = {'mag': odd_numbers} # array_dict라는 딕셔너리 생성 후 mag이라는 키로 odd_numbers 리스트를 저장
mag_series = pd.Series(array_dict['mag']) #딕셔너리에서 mag 키에 해당하는 값을 꺼내 pandas.Series 객체로 변환
print(mag_series)
```
